```python
df = df[["coaster_name", "Location", "Status",
         "Manufacturer", "year_introduced", "latitude",
         "longitude", "Type_Main", "opening_date_clean",
         "speed_mph", "height_ft", "Inversions_clean", "Gforce_clean"]].copy()
```

```python
df["opening_date_clean"] = pd.to_datetime(df["opening_date_clean"])
```

```python
df = df.rename(columns={
    "coaster_name":"Coaster_Name", "year_introduced":"Year_Introduced",
    "opening_date_clean":"Opening_Date", "speed_mph":"Speed_mph", 
    "height_ft":"Height_ft", "Inversions_clean":"Inversions", 
    "Gforce_clean":"Gforce"
})
```
